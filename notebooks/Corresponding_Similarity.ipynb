## Calcul de la Similarité entre un Mot d'Entrée et les Pays de la Liste

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
from transformers import RobertaTokenizer, RobertaModel

model_name = "roberta-base"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = RobertaModel.from_pretrained(model_name)
```

```python
def word_embedding(input_text):
    input_ids = tokenizer.encode(input_text, return_tensors="pt")
    with torch.no_grad():
        last_hidden_states = model(input_ids).last_hidden_state
    return last_hidden_states.mean(dim=1)[0]
```

```python
def calculate_word_embeddings(countries):

    # Calculer word embedding du country et region
    for country in countries:
        countrie = country['country.name'].lower()
        region   = country['region_name'].lower()
        country_embedding = word_embedding(countrie)
        region_embedding = word_embedding(region)
        country['country_embedding'] = country_embedding
        country['region_embedding'] = region_embedding

    return countries
```

```python
countries = calculate_word_embeddings(countries)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def similarities(input_word):
    # Initialiser la similarité maximale et le mot correspondant
    max_similarity = -1
    most_similar_word = None
    id = None

    # Obtenir l'embedding du mot d'entrée
    input_word = input_word.lower()
    input_embedding = word_embedding(input_word)

    # Parcourir chaque pays dans la liste
    for country in countries:
        # Calculer la similarité cosinus entre le mot d'entrée et le pays actuel
        similarity = cosine_similarity([input_embedding], [country['country_embedding']])

        # Mettre à jour la similarité maximale et le mot correspondant si la similarité actuelle est plus grande
        if similarity > max_similarity:
            max_similarity = similarity
            most_similar_word = country["country.name"]
            id = country["id"]

    return max_similarity, most_similar_word, id
```
